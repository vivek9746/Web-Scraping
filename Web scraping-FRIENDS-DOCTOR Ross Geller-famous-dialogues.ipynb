```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('http://www.glamourmagazine.co.uk/gallery/ross-geller-friends-best-quotes')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('p', attrs={'class':'bb-p'})
```

```python
result=results[3:]
```

```python
for r in range(len(result)):
    first=result[r]
    
    raw_names=first.find_all('strong')
    filt_names=[]
    for i in raw_names:
        filt=i.text[:-1]
        filt_names.append(filt)

    rawspt=[]
    a=first.contents
    for i in range(len(a)):
        if isinstance(a[i], str):
            rawspt.append(a[i])
    k=[]

    for i in range(len(rawspt)):
        if('[' and ']' in list(rawspt[i]) and r==0):
            new=[]
            new.append(i)
            rawspt[i]=rawspt[i]+rawspt[i+1]
            k.append(i+1)
    for l in k:
        del rawspt[l]

    del k[:]

    for x,y in zip(filt_names,rawspt):
        print(str(x)+" : "+y)
    print('\n\n')
    
```
